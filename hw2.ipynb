## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def beauty_derivative_plot(x_point):
    """
    Строим график функции f(x) = 2sin(x) - cos(2x), её производной
    и касательной в точке x_point
    """
    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)
    def f_derivative(x):
        return 2 * np.cos(x) + 2 * np.sin(2 * x)

    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = f(x)
    y_derivative = f_derivative(x)
    y_point = f(x_point)
    slope_point = f_derivative(x_point)

    tangent_line = y_point + slope_point * (x - x_point) # ур-е касательной

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8)) # два подграфика друг под другом

    ax1.plot(x, y, 'b-', linewidth=2, label='f(x) = 2sin(x) - cos(2x)')
    ax1.plot(x, tangent_line, 'r--', linewidth=2, label=f'Касательная в точке x = {x_point:.2f}')
    ax1.plot(x_point, y_point, 'ro', markersize=8, label=f'Точка касания ({x_point:.2f}, {y_point:.2f})')
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('y', fontsize=12)
    ax1.set_title('Функция и касательная', fontsize=14, fontweight='bold')
    ax1.legend(fontsize=10)
    ax1.grid(True, alpha=0.3)
    ax1.axhline(y=0, color='k', linewidth=0.5, alpha=0.5)
    ax1.axvline(x=0, color='k', linewidth=0.5, alpha=0.5)

    ax2.plot(x, y_derivative, 'g-', linewidth=2, label="f'(x) = 2cos(x) + 2sin(2x)")
    ax2.plot(x_point, slope_point, 'ro', markersize=8, label=f'Значение производной в точке: {slope_point:.2f}')
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel("f'(x)", fontsize=12)
    ax2.set_title('Производная', fontsize=14, fontweight='bold')
    ax2.legend(fontsize=10)
    ax2.grid(True, alpha=0.3)
    ax2.axhline(y=0, color='k', linewidth=0.5, alpha=0.5)
    ax2.axvline(x=0, color='k', linewidth=0.5, alpha=0.5)

    plt.tight_layout()
    plt.show()

    print(f"\nДля точки x = {x_point:.4f}")
    print(f"f(x) = {y_point:.4f}")
    print(f"f'(x) = {slope_point:.4f}")
````

## Problem 2

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
data = np.loadtxt('populations.txt')
years = data[:, 0].astype(int)
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

print(f"Годы: {years[0]} - {years[-1]}")
print(f"Форма данных: {data.shape}")
````

````python

plt.figure(figsize=(12, 6))
plt.plot(years, hares, 'b-', label='Hares', linewidth=2)
plt.plot(years, lynxes, 'r-', label='Lynxes', linewidth=2)
plt.plot(years, carrots, 'g-', label='Carrots', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Species Over Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
max_hare_year = years[np.argmax(hares)]
max_lynx_year = years[np.argmax(lynxes)]
max_carrot_year = years[np.argmax(carrots)]

print(f"Максимальная популяция зайцев: {max_hare_year} год")
print(f"Максимальная популяция рысей: {max_lynx_year} год")
print(f"Максимальная популяция моркови: {max_carrot_year} год")


plt.figure(figsize=(12, 6))
plt.plot(years, hares, 'b-', label='Hares', linewidth=2)
plt.plot(years, lynxes, 'r-', label='Lynxes', linewidth=2)
plt.plot(years, carrots, 'g-', label='Carrots', linewidth=2)

plt.plot(max_hare_year, hares.max(), 'b*', markersize=15, label=f'Hares max ({max_hare_year})')
plt.plot(max_lynx_year, lynxes.max(), 'r*', markersize=15, label=f'Lynxes max ({max_lynx_year})')
plt.plot(max_carrot_year, carrots.max(), 'g*', markersize=15, label=f'Carrots max ({max_carrot_year})')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Species Over Time with Maximum Points')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python
total_hare_lynx = hares + lynxes
hare_percentage = hares / total_hare_lynx * 100
lynx_percentage = lynxes / total_hare_lynx * 100

plt.figure(figsize=(14, 6))
plt.bar(years, hare_percentage, label='Hares', color='blue', alpha=0.7)
plt.bar(years, lynx_percentage, bottom=hare_percentage, label='Lynxes', color='red', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Percentage (%)')
plt.title('Relative Population of Hares and Lynxes (Stacked Bar Chart)')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.show()
````

````python
species_names = ['Hares', 'Lynxes', 'Carrots']
populations = [hares, lynxes, carrots]
colors = ['blue', 'red', 'green']

for i, (name, pop, color) in enumerate(zip(species_names, populations, colors)):
    mean_val = np.mean(pop)
    std_val = np.std(pop)
    print(f"{name}: среднее = {mean_val:.0f}, стандартное отклонение = {std_val:.0f}")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, (name, pop, color) in enumerate(zip(species_names, populations, colors)):
    axes[i].hist(pop, bins=8, color=color, alpha=0.7, edgecolor='black')
    mean_val = np.mean(pop)
    axes[i].axvline(mean_val, color='black', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.0f}')
    axes[i].set_xlabel('Population')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'{name} Population Distribution')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
hare_dominant = hares > lynxes
result_array = np.where(hare_dominant, 'H', 'L')
print("Результат:", result_array)

h_count = np.sum(hare_dominant)
l_count = len(hare_dominant) - h_count

plt.figure(figsize=(6, 6))
plt.pie([h_count, l_count], labels=['Hares dominant', 'Lynxes dominant'],
        colors=['blue', 'red'], autopct='%1.1f%%', startangle=90)
plt.title('Ratio of Years with Hares vs Lynxes Dominant')
plt.axis('equal')
plt.show()
````

````python
hare_change = np.diff(hares)
lynx_change = np.diff(lynxes)
change_years = years[1:]

plt.figure(figsize=(12, 6))
plt.plot(change_years, hare_change, 'b-', label='Hare population change', linewidth=2)
plt.plot(change_years, lynx_change, 'r-', label='Lynx population change', linewidth=2)
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.xlabel('Year')
plt.ylabel('Population Change')
plt.title('Year-to-Year Population Changes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


correlation = np.corrcoef(hare_change, lynx_change)[0, 1]
print(f"Коэффициент корреляции между изменениями популяций: {correlation:.3f}")
````

````python
plt.figure(figsize=(8, 8))
plt.scatter(hares, lynxes, c=years, cmap='viridis', s=100, alpha=0.7)
plt.colorbar(label='Year')
plt.xlabel('Hare Population')
plt.ylabel('Lynx Population')
plt.title('Hares vs Lynxes Population Scatter Plot')
plt.grid(True, alpha=0.3)


for i, year in enumerate(years):
    plt.annotate(str(year), (hares[i], lynxes[i]), fontsize=8, alpha=0.7)

plt.show()
````

````python
years_without_1920 = years[:-1]
hares_without_1920 = hares[:-1]
true_value_1920 = hares[-1]

from scipy import interpolate


last_years = years_without_1920[-3:]
last_hares = hares_without_1920[-3:]

f_linear = interpolate.interp1d(last_years, last_hares, kind='linear', fill_value='extrapolate')
estimated_linear = f_linear(1920)

f_poly = interpolate.interp1d(last_years, last_hares, kind='quadratic', fill_value='extrapolate')
estimated_poly = f_poly(1920)

estimated_mean = np.mean(last_hares)

print(f"Истинное значение популяции зайцев в 1920 году: {true_value_1920}")
print(f"Оценка (линейная интерполяция): {estimated_linear:.0f}, ошибка: {abs(estimated_linear - true_value_1920)/true_value_1920*100:.1f}%")
print(f"Оценка (полиномиальная): {estimated_poly:.0f}, ошибка: {abs(estimated_poly - true_value_1920)/true_value_1920*100:.1f}%")
print(f"Оценка (среднее за 3 года): {estimated_mean:.0f}, ошибка: {abs(estimated_mean - true_value_1920)/true_value_1920*100:.1f}%")


plt.figure(figsize=(10, 6))
plt.plot(years_without_1920, hares_without_1920, 'b-', label='Historical data', linewidth=2)
plt.plot(1920, true_value_1920, 'go', markersize=10, label='True value 1920')


plt.plot(1920, estimated_linear, 'r*', markersize=12, label='Linear extrapolation')
plt.plot(1920, estimated_poly, 'm*', markersize=12, label='Quadratic extrapolation')
plt.plot(1920, estimated_mean, 'c*', markersize=12, label='3-year average')

plt.xlabel('Year')
plt.ylabel('Hare Population')
plt.title('Hare Population Prediction for 1920')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````
